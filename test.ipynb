````python
def clean_model_json(raw_text: str) -> str:
    """
    Remove markdown code fences like ```json or ``` and return raw JSON.
    """
    cleaned = raw_text.strip()

    # Remove ```json ... ```
    if cleaned.startswith("```"):
        cleaned = cleaned.replace("```json", "").replace("```", "").strip()

    return cleaned
````
